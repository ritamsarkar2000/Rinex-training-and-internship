```python
#Analysis

#top 5 artist

top5a = df.groupby("artist_name").count().sort_values(by = "track_name", ascending = False)["track_name"][:5]
top5a
```

```python
top5a.plot.barh() #plotting the data
plt.show()
```

```python
#top 5 loudness track

top5loud = df[["loudness","track_name"]].sort_values(by = "loudness", ascending = True)[:5] #in case of loudness negative means louder
top5loud
```

```python
#plotting the data

plt.figure(figsize=(40,10))
sns.barplot(x = 'loudness', y = 'track_name', data = top5loud)
plt.title('top loudest 5 track')
plt.rcParams.update({'font.size': 25})  #to change the fornt size
plt.show()
```

```python
#artists with most dancing songs

top5_dance_track_artist = df[["danceability", "track_name", "artist_name"]].sort_values(by="danceability", ascending = False)[:5]
top5_dance_track_artist
```

```python
#plotting the data

plt.figure(figsize=(10,5))
sns.barplot(x = 'danceability', y = 'artist_name', data = top5_dance_track_artist)
plt.title("artists with most dancing songs")
plt.show()
```

```python
#top instrumentalness track

top = df[["instrumentalness", "track_name", "artist_name"]].sort_values(by="instrumentalness", ascending = False)[:5]
top
```

```python
#plotting the data

plt.figure(figsize=(7,5))
sns.barplot(x = 'instrumentalness', y = 'track_name', data = top)
plt.title("top instrumental tracks")
plt.show()

plt.figure(figsize=(10,10))
plt.pie(x = "instrumentalness", data=top, autopct = '1.2f%%', labels = top.track_name)
plt.show()
```

```python
#longest tracks

top = df[["duration_ms", "track_name", "artist_name"]].sort_values(by="duration_ms", ascending = False)[:5]
top
```

```python
#most common keys

top = df['key'].value_counts()[:5].index.tolist()
top
```
